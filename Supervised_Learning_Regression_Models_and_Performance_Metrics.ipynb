Q- 6 What is Logistic Regression? How does it differ from Linear Regression?

  ANS-- Logistic Regression â€” Definition

Logistic Regression is a statistical method used to model the relationship between one or more independent variables (predictors) and a categorical dependent variable, most commonly binary (having two possible outcomes, e.g., yes/no, pass/fail, disease/no disease).

Instead of predicting a continuous value like linear regression, logistic regression predicts the probability that an observation belongs to a particular class.


| **Aspect**                 | **Linear Regression**                             | **Logistic Regression**                                         |
| -------------------------- | ------------------------------------------------- | --------------------------------------------------------------- |
| **Dependent Variable (Y)** | Continuous (e.g., height, weight, income)         | Categorical (usually binary, e.g., 0/1)                         |
| **Output / Prediction**    | Predicts a numeric value of ( Y )                 | Predicts a probability (between 0 and 1)                        |
| **Model Equation**         | ( Y = \beta_0 + \beta_1 X + \varepsilon )         | ( \ln\left(\frac{p}{1 - p}\right) = \beta_0 + \beta_1 X )       |
| **Assumed Relationship**   | Linear relationship between ( X ) and ( Y )       | Linear relationship between ( X ) and the **log-odds** of ( Y ) |
| **Error Distribution**     | Errors (( \varepsilon )) are normally distributed | Follows a **binomial distribution**                             |
| **Predicted Values Range** | Can be any real number                            | Restricted to the range [0, 1]                                  |
| **Use Cases**              | Predicting continuous outcomes                    | Classifying categorical outcomes (e.g., spam vs. not spam)      |

Q-9 Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

(Include your Python code and output in the code box below.)



```python
# Import required libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
# X = independent variable (e.g., study hours)
# y = dependent variable (e.g., test scores)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape to 2D for scikit-learn
y = np.array([2, 4, 5, 4, 5])

# Create and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Print the slope (coefficient) and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
